```
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
samples = np.random.normal(loc=10, scale=2, size=1000)
sns.distplot(samples)
```

```
s = samples[samples > 0]
new = s + 5
sns.distplot(new)
```

```
from scipy.stats import skew # 偏度，找到非線性的圖形，偏移了多少
new = s ** 4
sns.distplot(new)
print("skew:", skew(new))
```

```
from scipy.stats import boxcox_normmax, boxcox
lamda = boxcox_normmax(new)
print(lamda)
new2 = boxcox(new, lamda)
print(skew(new2))
sns.distplot(new2)
```

```
new = s ** 0.1
sns.distplot(new)
print("skew:", skew(new))
```

```
lamda = boxcox_normmax(new)
print(lamda)
new2 = boxcox(new, lamda)
print(skew(new2))
sns.distplot(new2)
```
